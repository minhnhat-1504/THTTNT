## 1. TỔNG QUAN
### 1.1. Trí tuệ nhân tạo trong game
#### 1.1.1. Vai trò của AI trong trò chơi chiến thuật
<div style="text-align: justify;">
&emsp;&emsp;Trong lĩnh vực Khoa học máy tính, trò chơi điện tử – đặc biệt là các trò chơi chiến thuật và logic luôn được xem là nơi lý tưởng để nghiên cứu và phát triển Trí tuệ nhân tạo (AI). Khác với các ứng dụng AI trong xử lý ảnh hay ngôn ngữ tự nhiên, AI trong game tập trung vào bài toán ra quyết định và lập kế hoạch chiến lược. ### 1.2. Trò chơi tổng bằng không (Zero-sum Game)
#### 1.2.1. Định nghĩa và bản chất
<div style="text-align: justify;">
&emsp;&emsp;Khái niệm "Trò chơi tổng bằng không" (Zero-sum Game) được sử dụng để mô tả một tình huống xung đột lợi ích tuyệt đối giữa các bên tham gia. Cụ thể, trong mô hình này, lợi ích đạt được của người chơi này chính là tổn thất tương ứng của người chơi kia. Nói cách khác, tổng giá trị lợi ích của tất cả người chơi tại mọi trạng thái kết thúc luôn bằng không (hoặc bằng một hằng số cố định). Về mặt toán học, nếu gọi hàm lợi ích (Utility function) của người chơi A là $U_A$ và của người chơi B là $U_B$, thì trong một trò chơi tổng bằng không hai người, ta luôn có phương trình: <br><br>
$$U_A(s) + U_B(s) = 0, \quad \forall s \in S$$
<br>
&emsp;&emsp;(Trong đó $s$ là một trạng thái bất kỳ của trò chơi). <br>
&emsp;&emsp;Điều này hàm ý rằng không có cơ chế "hợp tác" trong loại hình trò chơi này. Mọi nỗ lực tối đa hóa điểm số của người chơi A (MAX) đồng nghĩa với việc đẩy người chơi B (MIN) về phía thua cuộc. Đây chính là cơ sở lý thuyết để xây dựng thuật toán Minimax: một bên cố gắng tìm nước đi để đạt giá trị cực đại, trong khi bên kia cố gắng chặn lại để giá trị đạt mức cực tiểu.
</div>

#### 1.2.2. Các đặc trưng cơ bản của môi trường trò chơi
<div style="text-align: justify;">
&emsp;&emsp;Để áp dụng thành công các thuật toán tìm kiếm đối kháng như Minimax hay Alpha-Beta, bài toán trò chơi cần thỏa mãn một số đặc trưng kỹ thuật cụ thể. Báo cáo này tập trung vào lớp bài toán có các tính chất sau:<br>
&emsp;&emsp;&emsp;1. Thông tin hoàn hảo (Perfect Information): Tại bất kỳ thời điểm nào, toàn bộ trạng thái của môi trường (bàn cờ, vị trí quân cờ) đều được công khai hoàn toàn cho cả hai người chơi. Không có yếu tố ẩn giấu như bài trên tay đối thủ trong Poker hay sương mù trong các game chiến thuật thời gian thực. Điều này cho phép thuật toán xây dựng được cây tìm kiếm chính xác mà không cần các yếu tố xác suất dự đoán thông tin ẩn.<br>
&emsp;&emsp;&emsp; 2. Môi trường tất định (Deterministic): Kết quả của một hành động được xác định duy nhất và không phụ thuộc vào yếu tố ngẫu nhiên (như gieo xúc xắc). Nếu tại trạng thái A, người chơi thực hiện nước đi $m$, trạng thái kế tiếp chắc chắn sẽ là B. Tính chất này đảm bảo rằng kế hoạch mà AI vạch ra sẽ không bị phá vỡ bởi "may mắn". <br>
&emsp;&emsp;&emsp;3. Lượt chơi luân phiên (Turn-based): Trò chơi diễn ra theo trình tự thời gian rời rạc, hai người chơi thay phiên nhau thực hiện hành động. Trạng thái của trò chơi sẽ tĩnh tại cho đến khi người chơi thực hiện nước đi của mình.
&emsp;&emsp;
<div style="text-align: justify;">

> Trong các trò chơi phức tạp như Cờ Caro, việc đi đến tận cùng ván cờ để nhận giá trị +1 hay -1 là bất khả thi về mặt tính toán. Do đó, khái niệm "tổng bằng không" được mở rộng sang các giá trị hàm đánh giá (Heuristic) tại các trạng thái trung gian, nơi điểm số dương thể hiện ưu thế nghiêng về MAX và điểm số âm thể hiện ưu thế nghiêng về MIN.
</div>
</div>

### 1.3. Các thành phần của mô hình trò chơi
<div>
&emsp;&emsp;Để chuyển đổi một trò chơi thực tế thành bài toán tính toán, ta mô hình hóa nó thông qua bốn thành phần cốt lõi, cùng nhau định nghĩa nên không gian trạng thái của bài toán:<br>
&emsp;&emsp;&emsp;1. Trạng thái khởi đầu ($S_0$): Là cấu hình ban đầu của bàn cờ (ví dụ: bàn cờ trống hoặc vị trí quân mặc định) cùng với việc xác định lượt đi đầu tiên. Đây là nút gốc của cây trò chơi.<br>
&emsp;&emsp;&emsp;2. Hàm sinh trạng thái kế tiếp (Successor Function): Xác định các quy tắc di chuyển hợp lệ. Với mọi trạng thái $s$, hàm này trả về tập hợp các cặp $(a, s')$, trong đó $a$ là nước đi và $s'$ là trạng thái kết quả.<br>
&emsp;&emsp;&emsp;3. Hàm kiểm tra kết thúc (Terminal Test): Một hàm Boolean xác định xem trạng thái hiện tại đã kết thúc trò chơi hay chưa (thắng, thua, hoặc hòa).<br>
&emsp;&emsp;&emsp;4. Hàm lợi ích (Utility Function): Gán một giá trị số học cho các trạng thái kết thúc (ví dụ: +1 cho thắng, -1 cho thua, 0 cho hòa) để định lượng kết quả phục vụ cho việc so sánh và tối ưu hóa.
</div>

### 1.4. Phân tích không gian trạng thái
<div>
&emsp;&emsp;Không gian tìm kiếm của các trò chơi đối kháng tăng trưởng theo hàm mũ $O(b^d)$ (với $b$ là hệ số nhánh và $d$ là độ sâu ván cờ). Trong các bài toán thực tế như Cờ vua hay Cờ tướng, số lượng trạng thái có thể lên tới $10^{123}$. Từ rào cản về kích thước không gian trên, việc áp dụng chiến lược duyệt vét cạn (Brute-force) để tìm ra kết quả thắng/thua tuyệt đối là bất khả thi về mặt thời gian và tài nguyên phần cứng.
</div>

## 2. Thuật toán Minimax


### 2.1. Nguyên lý hoạt động của Minimax

#### 2.1.1. Cơ chế
Minimax là thuật toán đệ quy được sử dụng để tìm ra quyết định tối ưu trong các trò chơi đối kháng (zero-sum games) có thông tin hoàn hảo. Nguyên lý cơ bản của thuật toán dựa trên giả định rằng **đối thủ cũng chơi tối ưu** (nghĩa là đối thủ luôn chọn nước đi gây bất lợi nhất cho ta).

Trong mô hình này, hai người chơi được định danh là:
* **MAX (Máy tính/AI):** Luôn cố gắng tìm nước đi để **tối đa hóa** (maximize) điểm lợi ích của mình.
* **MIN (Đối thủ):** Luôn cố gắng tìm nước đi để **tối thiểu hóa** (minimize) điểm lợi ích của MAX (tức là làm cho MAX thua hoặc đạt điểm thấp nhất).


#### 2.1.2. Chiến lược "Nhìn trước"
Minimax không chỉ xét nước đi hiện tại mà còn mô phỏng tương lai bằng cách duyệt sâu xuống cây trò chơi (Game Tree).
* Thuật toán sẽ thám hiểm cây trò chơi đến một độ sâu nhất định hoặc đến trạng thái kết thúc (Terminal state).
* Tại các nút lá, hàm đánh giá (Evaluation function) hoặc hàm lợi ích (Utility function) sẽ gán điểm số cho trạng thái đó.
* Giá trị này sau đó được lan truyền ngược (back-propagation) từ lá lên gốc để giúp MAX đưa ra quyết định tại trạng thái hiện tại.



### 2.2. Biểu diễn toán học của Minimax

Để hình thức hóa thuật toán, ta sử dụng các ký hiệu toán học cho các thành phần của cây trò chơi.

Gọi:
* $s$: Trạng thái hiện tại của bàn cờ.
* $PLAYER(s)$: Hàm xác định người chơi tại trạng thái $s$ (MAX hoặc MIN).
* $ACTIONS(s)$: Tập hợp các nước đi hợp lệ từ trạng thái $s$.
* $RESULT(s, a)$: Trạng thái kế tiếp sau khi thực hiện hành động $a$ lên trạng thái $s$.
* $UTILITY(s)$: Hàm lợi ích trả về giá trị số học tại trạng thái kết thúc.

Giá trị Minimax của một trạng thái $s$, ký hiệu là $V(s)$ hoặc $Minimax(s)$, được định nghĩa đệ quy như sau:

$$
V(s) =
\begin{cases}
UTILITY(s) & \text{nếu } TERMINAL\_TEST(s) \text{ là true} \\
\max_{a \in ACTIONS(s)} V(RESULT(s, a)) & \text{nếu } PLAYER(s) = MAX \\
\min_{a \in ACTIONS(s)} V(RESULT(s, a)) & \text{nếu } PLAYER(s) = MIN
\end{cases}
$$

**Giải thích công thức:**
1.  **Trường hợp cơ sở (Base case):** Nếu $s$ là trạng thái kết thúc, giá trị của nó chính là kết quả thắng/thua/hòa (ví dụ: +1, -1, 0).
2.  **Lượt của MAX:** MAX sẽ xét tất cả các trạng thái con $s'$ (được tạo ra từ các hành động $a$) và chọn giá trị **lớn nhất** trong số đó.
3.  **Lượt của MIN:** MIN sẽ xét tất cả các trạng thái con $s'$ và chọn giá trị **nhỏ nhất** trong số đó (để dìm MAX xuống).



### 2.3. Quy trình thực hiện

Quy trình thực thi của thuật toán Minimax đi theo mô hình tìm kiếm theo chiều sâu (Depth-First Search - DFS) bao gồm 5 bước chính:

**Bước 1: Khởi tạo và Tạo cây**
* Bắt đầu từ trạng thái hiện tại (nút gốc).
* Sinh ra các trạng thái con hợp lệ từ nút gốc. Tiếp tục quá trình này cho đến khi đạt đến độ sâu giới hạn hoặc trạng thái kết thúc game.

**Bước 2: Đánh giá tại nút lá**
* Khi thuật toán chạm tới đáy của cây tìm kiếm (nút lá), nó áp dụng hàm đánh giá (Heuristic hoặc Utility) để tính điểm số cho trạng thái đó.
* Ví dụ trong Tic-Tac-Toe: Nếu MAX thắng gán +1, MIN thắng gán -1, Hòa gán 0.

**Bước 3: Lan truyền ngược**
* Giá trị từ các nút lá được đẩy ngược lên nút cha trực tiếp của nó.

**Bước 4: Chọn lọc giá trị**
* Tại mỗi nút cha, thuật toán so sánh các giá trị nhận được từ các nút con:
    * Nếu nút cha là **MAX**: Nó lấy giá trị lớn nhất ($v = \max(v, \text{child\_value})$).
    * Nếu nút cha là **MIN**: Nó lấy giá trị nhỏ nhất ($v = \min(v, \text{child\_value})$).

**Bước 5: Ra quyết định**
* Quá trình lan truyền ngược tiếp tục cho đến khi quay về nút gốc.
* Tại nút gốc (lượt của MAX), AI sẽ chọn **hành động $a$** dẫn đến nút con có giá trị Minimax cao nhất.

**Ví dụ minh họa quy trình:**
Xét một cây trò chơi đơn giản :
1.  Tại các nút lá D, E, F có giá trị lần lượt là 2, 3, 4 (Giả sử đây là lượt MAX đánh giá).
2.  Nút cha của chúng là B (lượt MIN). MIN sẽ chọn $\min(2, 3, 4) = 2$. Vậy giá trị của nút B là 2.
3.  Tương tự, nếu nhánh bên kia nút C có giá trị là 5.
4.  Tại nút gốc A (lượt MAX), MAX so sánh B (2) và C (5). MAX chọn $\max(2, 5) = 5$.
5.  Kết luận: AI chọn đi theo nhánh dẫn đến C.

### 2.4. Phát biểu chi tiết bài toán
Bài toán đặt ra là xây dựng một tác tử thông minh (AI Agent) có khả năng tham gia trò chơi **Tic-Tac-Toe** (hoặc mở rộng ra Cờ Caro/Gomoku) dưới vai trò là một người chơi đấu với con người hoặc một AI khác.

  * **Bối cảnh:** Trò chơi diễn ra trên một bàn cờ kích thước $N \times N$ (với Tic-Tac-Toe là $3 \times 3$, Caro là $10 \times 10$ hoặc lớn hơn).
  * **Mục tiêu của AI:** Tìm kiếm chiến lược đi quân sao cho đảm bảo kết quả tốt nhất có thể: **Thắng** (nếu đối thủ sơ hở) hoặc **Hòa** (nếu đối thủ chơi tối ưu), tuyệt đối không để Thua.




### 2.5. Các thành phần của mô hình bài toán

Theo lý thuyết tìm kiếm đối kháng, bài toán được định nghĩa hình thức thông qua 5 thành phần sau:

1.  **Trạng thái ban đầu ($S_0$):** Bàn cờ rỗng kích thước $3 \times 3$, quy định người chơi X được đi trước.
2.  **Người chơi ($PLAYER(s)$):** Hàm xác định tại trạng thái $s$ hiện tại, lượt đi thuộc về ai (MAX hay MIN).
3.  **Hành động ($ACTIONS(s)$):** Tập hợp các nước đi hợp lệ $(i, j)$ mà người chơi có thể thực hiện tại trạng thái $s$ (là các ô còn trống).
4.  **Mô hình chuyển trạng thái ($RESULT(s, a)$):** Hàm trả về trạng thái bàn cờ mới sau khi thực hiện hành động $a$ lên trạng thái $s$.
5.  **Hàm kiểm tra kết thúc ($TERMINAL(s)$):** Hàm Boolean trả về `True` nếu trò chơi kết thúc (có hàng/cột/chéo đủ quân hoặc hết ô trống), ngược lại `False`.



### 2.6. Thiết kế và lựa chọn cấu trúc dữ liệu

#### 2.6.1. Biểu diễn bàn cờ (State Representation)

Việc lựa chọn cấu trúc dữ liệu ảnh hưởng trực tiếp đến tốc độ sinh nước đi và kiểm tra kết quả.

  * **Lựa chọn:** Sử dụng **Mảng 2 chiều (2D Array/Matrix)** hoặc **Danh sách lồng nhau (List of Lists)**.
      * Ví dụ: `board[3][3]`
  * **Lý giải:**
      * *Trực quan:* Ánh xạ 1-1 với bàn cờ thực tế, dễ dàng truy xuất qua tọa độ dòng `i` và cột `j`.
      * *Hiệu quả kiểm tra:* Dễ dàng duyệt vòng lặp để kiểm tra thắng thua theo hàng ngang, dọc và đường chéo.
      * *Tính bất biến:* Khi truyền vào đệ quy, cần sử dụng cơ chế sao chép sâu (`deepcopy`) để đảm bảo trạng thái cha không bị ảnh hưởng bởi phép thử ở trạng thái con.

#### 2.6.2. Biểu diễn giá trị ô cờ

  * **Lựa chọn:** Sử dụng chuỗi ký tự `"X"`, `"O"` và `None` (hoặc hằng số `EMPTY`).
  * **Lý giải:** Phù hợp với tư duy lập trình hướng đối tượng và dễ hiển thị ra giao diện (Console/UI). Trong các cài đặt tối ưu hơn (như C++), có thể dùng số nguyên (1, -1, 0) để tính tổng hàng/cột nhanh hơn, nhưng với Python, chuỗi ký tự là lựa chọn tốt nhất.



### 2.7. Đầu vào và Đầu ra của thuật toán

  * **Đầu vào (Input):**
      * `state`: Trạng thái bàn cờ hiện tại (một ma trận chứa các quân X, O, Empty).
      * `depth` (Tùy chọn): Độ sâu giới hạn nếu áp dụng cắt tỉa hoặc bàn cờ lớn.
      * `alpha`, `beta` (Nếu dùng Alpha-Beta): Các giá trị giới hạn để cắt tỉa nhánh.
  * **Đầu ra (Output):**
      * `best_move`: Tọa độ `(row, col)` của nước đi tối ưu nhất.
      * `value`: Giá trị điểm số dự kiến (ví dụ: +1 nếu chắc thắng, 0 nếu hòa).



### 2.8. Thiết kế Hàm đánh giá (Utility & Heuristic Function)

#### Trường hợp duyệt đến cùng (Tic-Tac-Toe 3x3)

Do không gian trạng thái nhỏ, ta có thể duyệt đến tận cùng (Terminal Node). Khi đó hàm đánh giá chính là **Hàm lợi ích (Utility Function)**:

  * Nếu $s$ là trạng thái MAX thắng: Trả về **+1** (hoặc +10).
  * Nếu $s$ là trạng thái MIN thắng: Trả về **-1** (hoặc -10).
  * Nếu $s$ là Hòa: Trả về **0**.

### 2.9. Mã giả thuật toán (Pseudocode)

#### Hàm quyết định nước đi (Minimax-Decision)

```plaintext
function MINIMAX-DECISION(state) returns an action
    v ← MAX-VALUE(state)
    return action trong SUCCESSORS(state) có giá trị v
```
#### Hàm tìm giá trị cực đại (Max-Value) - Dành cho lượt AI

```plaintext
function MAX-VALUE(state) returns a utility value
    if TERMINAL-TEST(state) then
        return UTILITY(state)
    
    v ← -∞
    for each (action, state') in SUCCESSORS(state) do
        v ← MAX(v, MIN-VALUE(state'))
    return v
```
#### Hàm tìm giá trị cực tiểu (Min-Value) - Dành cho lượt Đối thủ

```plaintext
function MIN-VALUE(state) returns a utility value
    if TERMINAL-TEST(state) then
        return UTILITY(state)
    
    v ← +∞
    for each (action, state') in SUCCESSORS(state) do
        v ← MIN(v, MAX-VALUE(state'))
    return v
```

**Triển khai thuật toán minimax vào trò chơi tictactoe 3x3 <br>
Bước 1: Khai báo thư viện và các hằng số cơ bản cho trò chơi.**

````python
import copy      # Thư viện dùng để tạo bản sao sâu (deep copy) của bàn cờ
import math      # Thư viện toán học (dùng cho giá trị vô cùng: inf)
import numpy as np # Thư viện hỗ trợ hiển thị ma trận

# Định nghĩa các hằng số đại diện cho quân cờ
X = "X"          # Đại diện cho người chơi MAX (đi trước)
O = "O"          # Đại diện cho người chơi MIN (đi sau)
EMPTY = None     # Đại diện cho ô trống trên bàn cờ

def initial_state():
    """
    Chức năng: Khởi tạo trạng thái ban đầu của bàn cờ.

    Tham số (Input):
        - Không có.

    Trả về (Output):
        - Một danh sách lồng nhau (list of lists) kích thước 3x3.
        - Tất cả các ô đều có giá trị là EMPTY (None).
    """
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]
````

**Bước 2: Định nghĩa các hàm xử lý luật chơi cơ bản: xác định lượt đi, tìm ô trống và thực hiện nước đi giả lập.**

````python
def player(board):
    """
    Chức năng: Xác định xem hiện tại là lượt đi của ai (X hay O).
    Quy tắc: X luôn đi trước.
             - Nếu số lượng X bằng số lượng O trên bàn cờ -> Đến lượt X.
             - Nếu số lượng X nhiều hơn O -> Đến lượt O.

    Tham số (Input):
        - board: Ma trận 3x3 trạng thái hiện tại.

    Trả về (Output):
        - Chuỗi "X" hoặc "O".
    """
    count_x = 0
    count_o = 0
    # Duyệt qua toàn bộ bàn cờ để đếm số quân
    for row in board:
        for cell in row:
            if cell == X:
                count_x += 1
            elif cell == O:
                count_o += 1

    if count_x > count_o:
        return O
    else:
        return X

def actions(board):
    """
    Chức năng: Tìm tất cả các nước đi hợp lệ (các ô còn trống).

    Tham số (Input):
        - board: Ma trận 3x3 trạng thái hiện tại.

    Trả về (Output):
        - Một tập hợp (set) chứa các tuple (hàng, cột).
        - Ví dụ: {(0, 0), (1, 2)}
    """
    res = set()
    board_len = len(board)
    for i in range(board_len):
        for j in range(board_len):
            if board[i][j] == EMPTY: # Nếu ô trống thì thêm vào danh sách
                res.add((i, j))
    return res

def result(board, action):
    """
    Chức năng: Trả về một bàn cờ MỚI sau khi thực hiện nước đi.
    Lưu ý: Phải dùng deepcopy để không làm thay đổi bàn cờ gốc đang xét.

    Tham số (Input):
        - board: Bàn cờ hiện tại.
        - action: Tuple (hàng, cột) vị trí muốn đánh.

    Trả về (Output):
        - Ma trận bàn cờ mới đã cập nhật nước đi.
    """
    # Kiểm tra nước đi có hợp lệ không
    (i, j) = action
    if board[i][j] is not EMPTY:
        raise Exception("Nước đi không hợp lệ: Ô đã có quân.")

    curr_player = player(board)         # Xác định người đánh
    result_board = copy.deepcopy(board) # Tạo bản sao độc lập
    result_board[i][j] = curr_player    # Cập nhật quân cờ
    return result_board
````

**Bước 3: Định nghĩa các hàm xác định người thắng, kẻ thua và gán điểm số cho các trạng thái kết thúc.**

````python
def winner(board):
    """
    Chức năng: Kiểm tra xem ai là người chiến thắng.

    Tham số (Input):
        - board: Bàn cờ hiện tại.

    Trả về (Output):
        - "X" nếu X thắng.
        - "O" nếu O thắng.
        - None nếu chưa ai thắng.
    """
    # 1. Kiểm tra 3 hàng ngang
    for i in range(3):
        if board[i][0] == board[i][1] == board[i][2] and board[i][0] is not None:
            return board[i][0]

    # 2. Kiểm tra 3 hàng dọc
    for j in range(3):
        if board[0][j] == board[1][j] == board[2][j] and board[0][j] is not None:
            return board[0][j]

    # 3. Kiểm tra 2 đường chéo
    if board[0][0] == board[1][1] == board[2][2] and board[0][0] is not None:
        return board[0][0]
    if board[0][2] == board[1][1] == board[2][0] and board[0][2] is not None:
        return board[0][2]

    return None

def terminal(board):
    """
    Chức năng: Kiểm tra trò chơi đã kết thúc chưa.

    Tham số (Input):
        - board: Bàn cờ hiện tại.

    Trả về (Output):
        - True: Nếu có người thắng HOẶC bàn cờ đã đầy (hòa).
        - False: Nếu trò chơi vẫn tiếp tục.
    """
    if winner(board) is not None:
        return True

    # Kiểm tra xem còn ô trống nào không
    for row in board:
        if EMPTY in row:
            return False
    return True

def utility(board):
    """
    Chức năng: Hàm Lợi ích (Utility Function) gán điểm cho trạng thái kết thúc.

    Tham số (Input):
        - board: Bàn cờ đã ở trạng thái kết thúc (terminal).

    Trả về (Output):
        -  1: Nếu X thắng (MAX thắng).
        - -1: Nếu O thắng (MIN thắng).
        -  0: Nếu Hòa.
    """
    win_player = winner(board)
    if win_player == X:
        return 1
    elif win_player == O:
        return -1
    else:
        return 0
````

**Bước 4: Định nghĩa giải thuật Minimax**

````python
def maxValue(state):
    """
    Chức năng: Tìm giá trị điểm số lớn nhất có thể cho lượt của MAX (X).
    Nguyên lý: Thử tất cả nước đi, giả định MIN sẽ chơi tối ưu để dìm điểm mình,
              nhưng mình sẽ chọn nhánh đỡ xấu nhất trong số đó.

    Tham số (Input): state (bàn cờ).
    Trả về (Output): Giá trị điểm số (v).
    """
    if terminal(state):
        return utility(state)

    v = -math.inf # Khởi tạo v là âm vô cùng (giá trị tồi nhất của MAX)

    for action in actions(state):
        # Gọi đệ quy minValue vì sau lượt mình là lượt đối thủ
        v = max(v, minValue(result(state, action)))

    return v

def minValue(state):
    """
    Chức năng: Tìm giá trị điểm số nhỏ nhất có thể cho lượt của MIN (O).
    Nguyên lý: MIN muốn điểm càng thấp càng tốt (-1).

    Tham số (Input): state (bàn cờ).
    Trả về (Output): Giá trị điểm số (v).
    """
    if terminal(state):
        return utility(state)

    v = math.inf # Khởi tạo v là dương vô cùng (giá trị tồi nhất của MIN)

    for action in actions(state):
        # Gọi đệ quy maxValue vì sau lượt mình là lượt đối thủ
        v = min(v, maxValue(result(state, action)))

    return v

def minimax(board):
    """
    Chức năng: Hàm quyết định nước đi tối ưu nhất cho người chơi hiện tại.

    Tham số (Input):
        - board: Bàn cờ hiện tại.

    Trả về (Output):
        - Tuple (hàng, cột) là nước đi tốt nhất.
        - Trả về None nếu game đã kết thúc.
    """
    if terminal(board):
        return None

    curr_player = player(board)

    # --- TRƯỜNG HỢP 1: Lượt của X (MAX) ---
    if curr_player == X:
        best_score = -math.inf
        best_move = None

        for action in actions(board):
            # Thử đi một nước, xem MIN sẽ trả lời thế nào (lấy giá trị minValue)
            score = minValue(result(board, action))

            # X muốn điểm cao nhất
            if score > best_score:
                best_score = score
                best_move = action
        return best_move

    # --- TRƯỜNG HỢP 2: Lượt của O (MIN) ---
    else:
        best_score = math.inf
        best_move = None

        for action in actions(board):
            # Thử đi một nước, xem MAX sẽ trả lời thế nào (lấy giá trị maxValue)
            score = maxValue(result(board, action))

            # O muốn điểm thấp nhất
            if score < best_score:
                best_score = score
                best_move = action
        return best_move
````

**Bước 5: Xây dựng chương trình chính**

````python
if __name__ == "__main__":
    # 1. Khởi tạo bàn cờ
    board = initial_state()

    print("------------------------------------------")
    print("   GAME TIC-TAC-TOE: NGƯỜI vs MÁY (AI)")
    print("------------------------------------------")

    # 2. Chọn phe
    while True:
        user_side = input("Bạn muốn chơi X hay O? (Lưu ý: X luôn đi trước): ").upper()
        if user_side in [X, O]:
            break
        print("Vui lòng chỉ nhập X hoặc O.")

    if user_side == X:
        ai_side = O
        print(f"-> Bạn chọn X. Bạn đi trước!")
    else:
        ai_side = X
        print(f"-> Bạn chọn O. Máy (X) đi trước!")

    # 3. Vòng lặp trò chơi
    while True:
        # Kiểm tra xem game kết thúc chưa
        game_over = terminal(board)
        curr_player = player(board)

        # Nếu kết thúc, in kết quả và thoát
        if game_over:
            print("\n=== TRÒ CHƠI KẾT THÚC ===")
            print(np.array(board))
            win = winner(board)
            if win is None:
                print(">> KẾT QUẢ: HÒA (Draw)!")
            elif win == user_side:
                print(">> KẾT QUẢ: BẠN THẮNG (Không thể xảy ra với Minimax chuẩn)!")
            else:
                print(">> KẾT QUẢ: MÁY THẮNG!")
            break

        # --- LƯỢT CỦA NGƯỜI CHƠI ---
        if curr_player == user_side:
            print(f"\n--- Lượt của BẠN ({user_side}) ---")
            print(np.array(board)) # In bàn cờ cho người dùng xem

            while True:
                try:
                    print("Nhập tọa độ nước đi (0-2):")
                    r = int(input("   Hàng: "))
                    c = int(input("   Cột : "))

                    if (r, c) in actions(board):
                        board = result(board, (r, c))
                        break
                    else:
                        print("!! Ô này không hợp lệ hoặc đã có người đánh. Thử lại.")
                except ValueError:
                    print("!! Vui lòng nhập số nguyên.")

        # --- LƯỢT CỦA MÁY (AI) ---
        else:
            print(f"\n--- Lượt của MÁY ({ai_side}) đang suy nghĩ... ---")
            move = minimax(board)
            board = result(board, move)
            print(f"Máy đã đánh vào ô: {move}")
````

## 3. Cắt tỉa Alpha-Beta cho bài toán tictactoe 5x5

### 3.1. Nguyên lý hoạt động
Cắt tỉa Alpha-Beta là một kỹ thuật tối ưu hóa cho thuật toán Minimax. Nó giúp giảm bớt số lượng nút cần duyệt trong cây trò chơi mà **không làm thay đổi kết quả cuối cùng** so với Minimax thuần túy.

* **Tư tưởng cốt lõi:** Nếu thuật toán tìm thấy một nước đi tồi tệ (đối với người chơi hiện tại) ở một nhánh, nó sẽ ngừng kiểm tra các nước đi còn lại của nhánh đó, vì chắc chắn nhánh đó sẽ không được chọn (hoặc đối thủ sẽ không cho phép chọn).
* **Hai tham số cắt tỉa:**
    * **Alpha ($\alpha$):** Giá trị tốt nhất (lớn nhất) mà người chơi **MAX** đã tìm thấy cho đến thời điểm hiện tại dọc theo đường đi. Giá trị $\alpha$ không bao giờ giảm.
    * **Beta ($\beta$):** Giá trị tốt nhất (nhỏ nhất) mà người chơi **MIN** đã tìm thấy cho đến thời điểm hiện tại dọc theo đường đi. Giá trị $\beta$ không bao giờ tăng.

### 3.2. Biểu diễn toán học và điều kiện cắt tỉa

#### Khởi tạo
Tại nút gốc, ta khởi tạo cửa sổ tìm kiếm với phạm vi rộng nhất:
* $\alpha = -\infty$
* $\beta = +\infty$


#### Quy tắc cập nhật
* **Tại nút MAX:** $\alpha = \max(\alpha, \text{giá trị của nút con})$.
* **Tại nút MIN:** $\beta = \min(\beta, \text{giá trị của nút con})$.

#### Điều kiện cắt tỉa
Việc cắt tỉa xảy ra ngay khi điều kiện sau được thỏa mãn:
$$\alpha \ge \beta$$

**Giải thích:**
* Nếu tại một nút MAX, ta tìm được một giá trị $\ge \beta$ (giá trị tốt nhất mà MIN đã có ở nhánh khác), thì MIN sẽ không bao giờ cho phép MAX đi vào nhánh này. Do đó, MAX không cần xét thêm các con còn lại.
* Tương tự, nếu tại một nút MIN, ta tìm được giá trị $\le \alpha$, MAX sẽ không bao giờ chọn đi vào nhánh này vì MAX đã có một lựa chọn tốt hơn ($\alpha$) ở chỗ khác.

### 3.3. Quy trình thực hiện chi tiết
Quy trình được thực hiện theo cơ chế duyệt chiều sâu (DFS) kết hợp cập nhật tham số :

1.  **Bước 1:** Bắt đầu tại nút gốc với khoảng $(\alpha=-\infty, \beta=+\infty)$.
2.  **Bước 2:** Duyệt xuống nút con đầu tiên bên trái, truyền giá trị $\alpha, \beta$ xuống.
3.  **Bước 3 (Tại nút lá):** Tính giá trị hàm lợi ích và trả ngược lên cha.
4.  **Bước 4 (Tại nút cha):**
    * Cập nhật giá trị $\alpha$ (nếu là MAX) hoặc $\beta$ (nếu là MIN) dựa trên giá trị con vừa trả về.
    * Kiểm tra điều kiện $\alpha \ge \beta$.
    * **Nếu thỏa:** Dừng việc duyệt các con còn lại của nút này (Cắt tỉa - Prune).
    * **Nếu không thỏa:** Tiếp tục duyệt nút con kế tiếp với giá trị $\alpha, \beta$ mới nhất.
5.  **Bước 5:** Lặp lại quá trình cho đến khi quay về nút gốc.

<br>

**Ví dụ minh họa:**
Giả sử tại nút MIN (C), ta đã tìm được một nhánh con có giá trị là 3. Lúc này $\beta$ tại C cập nhật thành 3 (hoặc nhỏ hơn). Trước đó, tại nút cha của C (là MAX, gọi là S), ta đã có $\alpha = 10$.
So sánh tại C: $\alpha (10) \ge \beta (3)$.
$\rightarrow$ Cắt tỉa! Không cần xét các con còn lại của C vì S (MAX) chắc chắn sẽ không chọn đường đi qua C (vì giá trị $\le 3$ thấp hơn 10) .

**Bước 1: Khai báo các thư viện cần thiết và định nghĩa các hằng số cơ bản.**



````python
import copy     # Thư viện để tạo bản sao độc lập của bàn cờ (Deep Copy)
import math     # Thư viện toán học (dùng cho vô cùng - math.inf)
import numpy as np # Thư viện để in bàn cờ dưới dạng ma trận cho đẹp

# --- ĐỊNH NGHĨA HẰNG SỐ ---
X = "X"          # Đại diện cho người chơi MAX (Máy)
O = "O"          # Đại diện cho người chơi MIN (Người)
EMPTY = None     # Đại diện cho ô trống

# --- CẤU HÌNH BÀN CỜ ---
N = 5            # Kích thước bàn cờ 5x5
WIN_COUNT = 4    # Số quân liên tiếp cần để thắng (Đặt là 4 cho game nhanh hơn)

def initial_state():
    """
    Chức năng: Khởi tạo trạng thái bắt đầu của bàn cờ NxN.

    Tham số: Không có.
    Trả về: Một danh sách lồng nhau (ma trận NxN) chứa toàn giá trị EMPTY.
    """
    return [[EMPTY for _ in range(N)] for _ in range(N)]
````

**Bước 2: Khai báo các hàm tìm nước đi và xác định lượt chơi. Hàm actions cần duyệt 25 ô.**



````python
def player(board):
    """
    Chức năng: Xác định lượt đi hiện tại là của ai (X hay O).
    Quy tắc: X đi trước.

    Tham số:
      - board: Trạng thái bàn cờ hiện tại.
    Trả về:
      - "X" hoặc "O".
    """
    count_x = 0
    count_o = 0
    for row in board:
        for cell in row:
            if cell == X:
                count_x += 1
            elif cell == O:
                count_o += 1

    # Nếu số lượng quân bằng nhau -> Lượt X (vì X đi trước).
    # Nếu X nhiều hơn -> Lượt O.
    if count_x > count_o:
        return O
    else:
        return X

def actions(board):
    """
    Chức năng: Tìm tất cả các nước đi hợp lệ (các ô còn trống).

    Tham số:
      - board: Bàn cờ hiện tại.
    Trả về:
      - Tập hợp (set) các tọa độ (hàng, cột) có thể đi.
    """
    res = set()
    board_len = len(board)
    for i in range(board_len):
        for j in range(board_len):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

def result(board, action):
    """
    Chức năng: Trả về bàn cờ MỚI sau khi thực hiện nước đi.
    Lưu ý: Sử dụng copy.deepcopy như yêu cầu của tài liệu để đảm bảo an toàn dữ liệu.

    Tham số:
      - board: Bàn cờ hiện tại.
      - action: Tọa độ (hàng, cột) muốn đánh.
    Trả về:
      - Bàn cờ mới với nước đi đã được cập nhật.
    """
    curr_player = player(board)         # Xác định ai đang đi
    result_board = copy.deepcopy(board) # Tạo bản sao sâu (tốn bộ nhớ nhưng an toàn)
    (i, j) = action

    # Kiểm tra tính hợp lệ
    if result_board[i][j] is not EMPTY:
        raise Exception("Nước đi không hợp lệ: Ô đã có quân.")

    result_board[i][j] = curr_player    # Đánh dấu quân cờ
    return result_board
````

**Bước 3**


````python
def check_line(board, player):
    """
    Hàm phụ trợ: Kiểm tra xem 'player' có đạt đủ số quân liên tiếp (WIN_COUNT) không.
    Quét theo 4 hướng: Ngang, Dọc, Chéo chính, Chéo phụ.
    """
    # 1. Kiểm tra hàng ngang
    for r in range(N):
        for c in range(N - WIN_COUNT + 1):
            if all(board[r][c+k] == player for k in range(WIN_COUNT)): return True

    # 2. Kiểm tra hàng dọc
    for c in range(N):
        for r in range(N - WIN_COUNT + 1):
            if all(board[r+k][c] == player for k in range(WIN_COUNT)): return True

    # 3. Kiểm tra chéo chính (\)
    for r in range(N - WIN_COUNT + 1):
        for c in range(N - WIN_COUNT + 1):
            if all(board[r+k][c+k] == player for k in range(WIN_COUNT)): return True

    # 4. Kiểm tra chéo phụ (/)
    for r in range(N - WIN_COUNT + 1):
        for c in range(WIN_COUNT - 1, N):
            if all(board[r+k][c-k] == player for k in range(WIN_COUNT)): return True

    return False

def winner(board):
    """
    Chức năng: Trả về người thắng cuộc.
    Trả về: "X", "O" hoặc None.
    """
    if check_line(board, X): return X
    if check_line(board, O): return O
    return None

def terminal(board):
    """
    Chức năng: Kiểm tra trò chơi kết thúc chưa (Có người thắng hoặc hết ô).
    Trả về: True / False.
    """
    if winner(board) is not None:
        return True

    # Kiểm tra còn ô trống không
    for row in board:
        if EMPTY in row:
            return False
    return True

def utility(board):
    """
    Chức năng: Hàm Lợi ích (Utility Function) trả về giá trị của trạng thái.

    Lưu ý quan trọng cho 5x5:
    - Nếu game kết thúc: Trả về điểm tuyệt đối (10000 hoặc -10000).
    - Nếu game chưa kết thúc (hết độ sâu tìm kiếm): Trả về điểm đánh giá (heuristic)
      dựa trên số lượng quân cờ trên bàn.

    Trả về: Số nguyên (điểm số).
    """
    win = winner(board)
    if win == X:
        return 10000   # MAX thắng -> Điểm rất cao
    elif win == O:
        return -10000  # MIN thắng -> Điểm rất thấp

    # --- PHẦN ĐÁNH GIÁ HEURISTIC (Cho trường hợp chưa kết thúc) ---
    # Đếm số lượng quân để ước lượng lợi thế
    score_x = 0
    score_o = 0

    # Quét toàn bàn cờ (Heuristic đơn giản: đếm số quân)
    # (Trong thực tế có thể phức tạp hơn như đếm chuỗi liên tiếp)
    for r in range(N):
        for c in range(N):
            if board[r][c] == X: score_x += 1
            elif board[r][c] == O: score_o += 1

    return score_x - score_o # Trả về hiệu số quân cờ (Lợi thế của X)
````

**Bước 4: Định nghĩa hàm cắt tỉa**

````python
# =============================================================================
# BLOCK 4: THUẬT TOÁN MINIMAX VỚI CẮT TỈA ALPHA-BETA
# =============================================================================

def maxValue(state, alpha, beta, depth):
    """
    Chức năng: Tìm giá trị lớn nhất cho MAX (X) có cắt tỉa.

    Tham số:
      - state: Bàn cờ hiện tại.
      - alpha: Giá trị tốt nhất hiện có của MAX.
      - beta: Giá trị tốt nhất hiện có của MIN.
      - depth: Độ sâu còn lại.
    """
    # Điều kiện dừng: Hết độ sâu hoặc Game kết thúc
    if depth == 0 or terminal(state):
        return utility(state)

    v = -math.inf

    # Duyệt qua các hành động
    for action in actions(state):
        # Gọi hàm minValue cho tầng tiếp theo (giảm độ sâu đi 1)
        v = max(v, minValue(result(state, action), alpha, beta, depth - 1))

        # Cắt tỉa Alpha-Beta:
        # Nếu giá trị tìm được (v) lớn hơn hoặc bằng kỳ vọng thấp nhất của MIN (beta)
        # -> MIN sẽ không bao giờ chọn nhánh này -> Cắt tỉa.
        if v >= beta:
            return v

        # Cập nhật giá trị tốt nhất của MAX
        alpha = max(alpha, v)

    return v

def minValue(state, alpha, beta, depth):
    """
    Chức năng: Tìm giá trị nhỏ nhất cho MIN (O) có cắt tỉa.
    """
    # Điều kiện dừng
    if depth == 0 or terminal(state):
        return utility(state)

    v = math.inf

    for action in actions(state):
        # Gọi hàm maxValue cho tầng tiếp theo
        v = min(v, maxValue(result(state, action), alpha, beta, depth - 1))

        # Cắt tỉa Alpha-Beta:
        # Nếu giá trị tìm được (v) nhỏ hơn hoặc bằng kỳ vọng cao nhất của MAX (alpha)
        # -> MAX sẽ không bao giờ chọn nhánh này -> Cắt tỉa.
        if v <= alpha:
            return v

        # Cập nhật giá trị tốt nhất của MIN
        beta = min(beta, v)

    return v

def minimax_alpha_beta(board, depth):
    """
    Chức năng: Hàm chính tìm nước đi tối ưu nhất cho người chơi hiện tại.

    Tham số:
      - board: Bàn cờ hiện tại.
      - depth: Độ sâu giới hạn tìm kiếm (ví dụ: 2 hoặc 3).

    Trả về:
      - Tọa độ (hàng, cột) của nước đi tốt nhất.
    """
    current_player = player(board)
    best_move = None

    # Nếu là lượt MAX (X) -> Gọi minValue để tìm max
    if current_player == X:
        v = -math.inf
        for action in actions(board):
            # Khởi tạo alpha = -vô cùng, beta = +vô cùng
            val = minValue(result(board, action), -math.inf, math.inf, depth - 1)
            if val > v:
                v = val
                best_move = action

    # Nếu là lượt MIN (O) -> Gọi maxValue để tìm min
    else:
        v = math.inf
        for action in actions(board):
            val = maxValue(result(board, action), -math.inf, math.inf, depth - 1)
            if val < v:
                v = val
                best_move = action

    return best_move
````

**Bước 5: Main**

````python
if __name__ == "__main__":
    # Khởi tạo bàn cờ
    board = initial_state()
    print(f"--- TIC-TAC-TOE {N}x{N} (ALPHA-BETA, DEEPCOPY) ---")
    print(f"Luật thắng: {WIN_COUNT} quân liên tiếp.")

    # Thiết lập: AI là X (đi trước)
    ai_turn = True
    print("AI đóng vai X. AI đi trước.")

    while True:
        # 1. In bàn cờ hiện tại
        print("\nBàn cờ hiện tại:")
        print(np.array(board))

        # 2. Kiểm tra kết thúc game
        if terminal(board):
            score = utility(board)
            if score >= 1000:
                print("\n>> KẾT QUẢ: AI (X) THẮNG!")
            elif score <= -1000:
                print("\n>> KẾT QUẢ: BẠN (O) THẮNG!")
            else:
                print("\n>> KẾT QUẢ: HÒA!")
            break

        # 3. Xử lý lượt đi
        if ai_turn:
            print("\nAI đang suy nghĩ (Độ sâu 2)...")
            # Lưu ý: Do dùng deepcopy nên rất chậm, chỉ nên để depth=2
            move = minimax_alpha_beta(board, depth=2)

            if move:
                board = result(board, move)
                print(f"AI đã đánh vào ô: {move}")
            else:
                print("AI không tìm thấy nước đi (Lỗi hoặc Hòa).")
                break
            ai_turn = False # Chuyển lượt cho người chơi

        else:
            # Lượt người chơi
            try:
                print("\nLượt của BẠN (O). Hãy nhập tọa độ:")
                r = int(input(f"   Nhập hàng (0-{N-1}): "))
                c = int(input(f"   Nhập cột (0-{N-1}): "))

                # Kiểm tra nước đi hợp lệ
                if (r, c) in actions(board):
                    board = result(board, (r, c))
                    ai_turn = True # Chuyển lượt cho AI
                else:
                    print("Lỗi: Nước đi không hợp lệ hoặc ô đã có quân.")
            except ValueError:
                print("Lỗi: Vui lòng nhập số nguyên.")
````
