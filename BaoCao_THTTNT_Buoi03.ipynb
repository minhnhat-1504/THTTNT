### CHƯƠNG 1: TỔNG QUAN
#### 1.1. Trí tuệ nhân tạo trong game
##### 1.1.1. Vai trò của AI trong trò chơi chiến thuật
<div style="text-align: justify;">
&emsp;&emsp;Trong lĩnh vực Khoa học máy tính, trò chơi điện tử – đặc biệt là các trò chơi chiến thuật và logic luôn được xem là nơi lý tưởng để nghiên cứu và phát triển Trí tuệ nhân tạo (AI). Khác với các ứng dụng AI trong xử lý ảnh hay ngôn ngữ tự nhiên, AI trong game tập trung vào bài toán ra quyết định và lập kế hoạch chiến lược. Mục tiêu cốt lõi là xây dựng các tác tử thông minh (intelligent agents) có khả năng tư duy, phân tích trạng thái bàn cờ và đưa ra các nước đi hợp lý nhằm đạt được mục tiêu thắng cuộc trước đối thủ.
<br>
&emsp;&emsp;Trong bài toán tìm kiếm đối kháng, AI không hoạt động trong môi trường tĩnh mà phải đối mặt với một thực thể thông minh khác. Do đó, thuật toán không chỉ đơn thuần là tìm đường đi ngắn nhất, mà là tìm kiếm một chiến lược tối ưu để ứng phó với mọi kịch bản phản kích từ đối thủ.
</div>

##### 1.1.2. Phân biệt tìm kiếm đơn tác tử và tìm kiếm đối kháng
<div style="text-align: justify;">
&emsp;&emsp;&emsp;&emsp;Trong các bài toán tìm kiếm đơn tác tử (như giải mê cung hay bài toán xếp hình 8-puzzle), tác tử hoạt động trong một môi trường tĩnh hoặc chỉ thay đổi theo quy luật tự nhiên, không có sự can thiệp chủ đích từ bên ngoài. Ngược lại, tìm kiếm đối kháng diễn ra trong môi trường cạnh tranh khốc liệt, nơi mỗi hành động của tác tử (MAX) sẽ dẫn đến một hành động đáp trả của đối thủ (MIN) nhằm giảm thiểu lợi ích của tác tử đó. <br>

| Tiêu chí so sánh | Tìm kiếm đơn tác tử | Tìm kiếm đối kháng |
| :--- | :--- | :--- |
| **Đại diện tiêu biểu** | Giải mê cung, Tìm đường đi, Xếp hình (Puzzle). | Cờ vua, Cờ tướng, Cờ Caro, Tic-Tac-Toe. |
| **Tính chất môi trường** | Tĩnh (Static): Môi trường chỉ thay đổi do tác động của chính tác tử. | Động & Cạnh tranh (Dynamic & Hostile): Môi trường thay đổi do cả tác tử và đối thủ. |
| **Mục tiêu thuật toán** | Tìm một chuỗi hành động (path) từ trạng thái đầu đến đích với chi phí thấp nhất. | Tìm một chiến lược (strategy) để tối đa hóa khả năng thắng lợi bất kể đối thủ đi như thế nào. |
| **Độ phức tạp** | Thấp hơn, thường chỉ phụ thuộc vào không gian trạng thái của bài toán. | Rất cao, phụ thuộc vào nước đi của đối thủ, dễ dẫn đến bùng nổ tổ hợp. |
| **Giả định** | Mọi hành động đều diễn ra đúng như dự kiến. | Phải giả định đối thủ luôn chơi tối ưu (luôn chọn nước đi gây bất lợi nhất cho ta). |
<div style="text-align: justify;">

> Nhận định: Có thể thấy, thách thức lớn nhất của tìm kiếm đối kháng chính là tính "bất định do đối thủ". Vì không thể biết chính xác đối phương sẽ đi nước nào, thuật toán buộc phải tính toán dựa trên giả định xấu nhất (Worst-case scenario analysis): "Nếu mình đi nước A, đối thủ giỏi nhất sẽ đáp trả thế nào?". Đây chính là cơ sở lý thuyết hình thành nên thuật toán Minimax và các kỹ thuật cắt tỉa sẽ được trình bày ở phần sau.
